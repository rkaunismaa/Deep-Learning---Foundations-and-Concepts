```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the parameters for the Gaussian distribution
mu = 0  # mean
sigma = 1  # standard deviation

# Define the x range
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)

# Calculate the Gaussian distribution
y = norm.pdf(x, mu, sigma)

# Plot the Gaussian distribution
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'r-', label='Gaussian Distribution')

# Add mean and standard deviation lines
plt.axvline(mu, color='black', linestyle='--')
plt.annotate(r'$\mu$', xy=(mu, 0), xytext=(mu, -0.02),
             textcoords='data', ha='center', fontsize=12)

# Add 2 sigma range
plt.plot([mu - sigma, mu + sigma], [0.1, 0.1], 'k-', lw=2)
plt.annotate(r'$2\sigma$', xy=(mu, 0.1), xytext=(mu, 0.12),
             textcoords='data', ha='center', fontsize=12, arrowprops=dict(arrowstyle='<->', lw=1.5))

# Labels and title
plt.xlabel('$x$')
plt.ylabel(r'$\mathcal{N}(x|\mu,\sigma^2)$')
plt.title('Figure 2.8: Gaussian distribution with mean $\mu$ and standard deviation $\sigma$')

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show the plot
plt.show()
```
