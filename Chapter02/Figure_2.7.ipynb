```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range
x = np.linspace(-2, 4, 1000)

# Uniform distribution
def uniform(x):
    return np.where((x >= -1) & (x <= 1), 0.5, 0)

# Exponential distribution
def exponential(x, lam=1):
    return lam * np.exp(-lam * x) * (x >= 0)

# Laplace distribution
def laplace(x, mu=1, gamma=1):
    return (1/(2*gamma)) * np.exp(-np.abs(x - mu) / gamma)

# Generate y values for each distribution
y_uniform = uniform(x)
y_exponential = exponential(x)
y_laplace = laplace(x)

# Plot the distributions
plt.figure(figsize=(8, 5))
plt.plot(x, y_uniform, 'r-', label='Uniform (-1, 1)')
plt.plot(x, y_exponential, 'b-', label='Exponential (λ=1)')
plt.plot(x, y_laplace, 'g-', label='Laplace (μ=1, γ=1)')

# Adding the lines at specific points for better visualization
plt.axvline(x=0, color='grey', linewidth=0.5)
plt.axvline(x=1, color='blue', linewidth=0.5)

# Labels and legend
plt.xlabel('x')
plt.ylabel('p(x)')
plt.ylim(0, 1.1)
plt.legend(loc='upper right')
plt.title('Figure 2.7: Plots of different distributions')
plt.show()
```
