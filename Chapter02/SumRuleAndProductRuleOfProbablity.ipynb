```python
# Probability Demonstrations with Sympy: Sum Rule and Product Rule

## Introduction
# In this notebook, we will demonstrate two fundamental rules of probability: the Sum Rule and the Product Rule using sympy for symbolic computation.

import sympy as sp
import matplotlib.pyplot as plt

## Sum Rule of Probability
# The sum rule states that the probability of the occurrence of at least one of two mutually exclusive events is the sum of their individual probabilities.
# Mathematically, for two mutually exclusive events A and B:
# P(A or B) = P(A) + P(B)

# Define symbolic variables for the probabilities
P_A, P_B = sp.symbols('P_A P_B')

# Sum rule expression
sum_rule = P_A + P_B

# Display the sum rule
sum_rule
```

```python
# Example: Rolling a Die
# Event A: Rolling a 2
# Event B: Rolling a 5
# Both events are mutually exclusive because a single roll cannot result in both a 2 and a 5.

P_A_val = 1/6  # Probability of rolling a 2
P_B_val = 1/6  # Probability of rolling a 5

# Calculate the probability of A or B
P_A_or_B_val = sum_rule.subs({P_A: P_A_val, P_B: P_B_val})

print(f"P(A or B) = P(A) + P(B) = {P_A_val} + {P_B_val} = {P_A_or_B_val}")

# Visualization
events = ['A', 'B']
probabilities = [P_A_val, P_B_val]

plt.bar(events, probabilities, color=['blue', 'orange'])
plt.title('Sum Rule of Probability')
plt.xlabel('Events')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.show()
```

```python
## Product Rule of Probability
# The product rule states that the probability of the occurrence of two independent events A and B is the product of their individual probabilities.
# Mathematically, for two independent events A and B:
# P(A and B) = P(A) * P(B)

# Product rule expression
product_rule = P_A * P_B

# Display the product rule
product_rule
```

```python
# Example: Flipping Two Coins
# Event A: Getting a Head on the first coin
# Event B: Getting a Head on the second coin
# Both events are independent because the outcome of one coin flip does not affect the other.

P_A_val = 1/2  # Probability of getting a Head on the first coin
P_B_val = 1/2  # Probability of getting a Head on the second coin

# Calculate the probability of A and B
P_A_and_B_val = product_rule.subs({P_A: P_A_val, P_B: P_B_val})

print(f"P(A and B) = P(A) * P(B) = {P_A_val} * {P_B_val} = {P_A_and_B_val}")

# Visualization
events = ['A', 'B']
probabilities = [P_A_val, P_B_val]

plt.bar(events, probabilities, color=['green', 'red'])
plt.title('Product Rule of Probability')
plt.xlabel('Events')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.show()
```

```python
# The probability of both events happening (A and B)
combined_event = ['A and B']
combined_probability = [P_A_and_B_val]

plt.bar(combined_event, combined_probability, color=['purple'])
plt.title('Combined Probability (Product Rule)')
plt.xlabel('Event')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.show()

## Conclusion
# The Sum Rule and the Product Rule are fundamental concepts in probability theory.
# We have demonstrated these rules with symbolic expressions and visualizations using Python and sympy.
```
