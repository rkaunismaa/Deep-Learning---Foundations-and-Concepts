```python
# Bayes' Theorem Demonstration

## Introduction
# In this notebook, we will demonstrate Bayes' Theorem using sympy for symbolic computation and some example scenarios to illustrate the concept.

import sympy as sp
import matplotlib.pyplot as plt

## Bayes' Theorem
# Bayes' Theorem describes the probability of an event, based on prior knowledge of conditions that might be related to the event.
# Mathematically, it is expressed as:
# P(A|B) = [P(B|A) * P(A)] / P(B)

# Define symbolic variables
P_A, P_B, P_B_given_A, P_A_given_B = sp.symbols('P_A P_B P_B_given_A P_A_given_B')

# Bayes' Theorem expression
bayes_theorem = sp.Eq(P_A_given_B, (P_B_given_A * P_A) / P_B)

# Display the theorem
bayes_theorem
```

```python
# Example: Medical Test
# Suppose we have a medical test for a disease.
# - P(D) is the probability of having the disease.
# - P(T|D) is the probability of testing positive given the presence of the disease.
# - P(T|~D) is the probability of testing positive given the absence of the disease.
# - P(~D) is the probability of not having the disease.
# We want to find the probability of having the disease given a positive test result, P(D|T).

# Given values
P_D_val = 0.01       # Probability of having the disease
P_T_given_D_val = 0.99  # Probability of testing positive given the disease
P_T_given_not_D_val = 0.05  # Probability of testing positive given no disease
P_not_D_val = 1 - P_D_val  # Probability of not having the disease

# Calculate P(T)
P_T_val = (P_T_given_D_val * P_D_val) + (P_T_given_not_D_val * P_not_D_val)

# Calculate P(D|T) using Bayes' Theorem
P_D_given_T_val = (P_T_given_D_val * P_D_val) / P_T_val

print(f"P(D) = {P_D_val}")
print(f"P(T|D) = {P_T_given_D_val}")
print(f"P(T|~D) = {P_T_given_not_D_val}")
print(f"P(~D) = {P_not_D_val}")
print(f"P(T) = {P_T_val}")
print(f"P(D|T) = {P_D_given_T_val}")
```

```python
# Visualization of the probabilities
labels = ['P(D)', 'P(T|D)', 'P(T|~D)', 'P(~D)', 'P(T)', 'P(D|T)']
values = [P_D_val, P_T_given_D_val, P_T_given_not_D_val, P_not_D_val, P_T_val, P_D_given_T_val]

plt.figure(figsize=(10, 5))
plt.bar(labels, values, color=['blue', 'orange', 'green', 'red', 'purple', 'brown'])
plt.title('Bayes\' Theorem Probabilities')
plt.xlabel('Probabilities')
plt.ylabel('Values')
plt.ylim(0, 1)
plt.show()

## Conclusion
# Bayes' Theorem is a powerful tool for calculating conditional probabilities.
# In this notebook, we have demonstrated the theorem with symbolic expressions using sympy and a practical example of a medical test scenario.
```
