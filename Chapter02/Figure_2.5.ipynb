```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
X = np.random.randint(1, 10, 60)  # X takes 9 possible values
Y = np.random.choice([1, 2], 60)  # Y takes 2 possible values

# Plot the distributions
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Joint distribution p(X, Y)
axes[0, 0].scatter(X, Y, color='blue')
axes[0, 0].set_title('$p(X, Y)$')
axes[0, 0].set_xticks(np.arange(1, 10))
axes[0, 0].set_yticks([1, 2])
axes[0, 0].set_xlabel('$X$')
axes[0, 0].set_ylabel('$Y$')
axes[0, 0].grid(True, which='both', color='red', linestyle='-', linewidth=1)
axes[0, 0].minorticks_on()
axes[0, 0].grid(True, which='minor', color='red', linestyle='-', linewidth=1)

# Marginal distribution p(Y)
axes[0, 1].hist(Y, bins=[0.5, 1.5, 2.5], edgecolor='black', color='blue', alpha=0.3, rwidth=0.6)
axes[0, 1].set_title('$p(Y)$')
axes[0, 1].set_xticks([1, 2])
axes[0, 1].set_xlabel('$Y$')
axes[0, 1].set_ylabel('Frequency')

# Marginal distribution p(X)
axes[1, 0].hist(X, bins=np.arange(1, 11) - 0.5, edgecolor='black', color='blue', alpha=0.3)
axes[1, 0].set_title('$p(X)$')
axes[1, 0].set_xticks(np.arange(1, 10))
axes[1, 0].set_xlabel('$X$')
axes[1, 0].set_ylabel('Frequency')

# Conditional distribution p(X|Y=1)
X_given_Y1 = X[Y == 1]
axes[1, 1].hist(X_given_Y1, bins=np.arange(1, 11) - 0.5, edgecolor='black', color='blue', alpha=0.3)
axes[1, 1].set_title('$p(X|Y=1)$')
axes[1, 1].set_xticks(np.arange(1, 10))
axes[1, 1].set_xlabel('$X$')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
