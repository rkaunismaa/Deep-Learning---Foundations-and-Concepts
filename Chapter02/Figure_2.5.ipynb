```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
X = np.random.randint(1, 10, 60)  # X takes 9 possible values
Y = np.random.choice([1, 2], 60)  # Y takes 2 possible values

# Plot the distributions
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Joint distribution p(X, Y)
axes[0, 0].scatter(X, Y, color='blue')
axes[0, 0].set_title('$p(X, Y)$')
axes[0, 0].set_xticks(np.arange(1, 10))
axes[0, 0].set_yticks([1, 2])
axes[0, 0].set_xlabel('$X$')
axes[0, 0].set_ylabel('$Y$')
axes[0, 0].grid(True, which='both', color='red', linestyle='-', linewidth=1)
axes[0, 0].minorticks_on()
axes[0, 0].grid(True, which='minor', color='red', linestyle='-', linewidth=1)

# Marginal distribution p(Y)
axes[0, 1].hist(Y, bins=[0.5, 1.5, 2.5], edgecolor='black', color='blue', alpha=0.3, rwidth=0.6)
axes[0, 1].set_title('$p(Y)$')
axes[0, 1].set_xticks([1, 2])
axes[0, 1].set_xlabel('$Y$')
axes[0, 1].set_ylabel('Frequency')

# Marginal distribution p(X)
axes[1, 0].hist(X, bins=np.arange(1, 11) - 0.5, edgecolor='black', color='blue', alpha=0.3)
axes[1, 0].set_title('$p(X)$')
axes[1, 0].set_xticks(np.arange(1, 10))
axes[1, 0].set_xlabel('$X$')
axes[1, 0].set_ylabel('Frequency')

# Conditional distribution p(X|Y=1)
X_given_Y1 = X[Y == 1]
axes[1, 1].hist(X_given_Y1, bins=np.arange(1, 11) - 0.5, edgecolor='black', color='blue', alpha=0.3)
axes[1, 1].set_title('$p(X|Y=1)$')
axes[1, 1].set_xticks(np.arange(1, 10))
axes[1, 1].set_xlabel('$X$')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
X = np.random.randint(0, 9, 60)
Y = np.random.randint(1, 3, 60)

# Create the figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Plot p(X,Y)
axs[0, 0].scatter(X, Y, alpha=0.7)
axs[0, 0].set_xlim(-0.5, 8.5)
axs[0, 0].set_ylim(0.5, 2.5)
axs[0, 0].set_yticks([1, 2])
axs[0, 0].set_yticklabels(['Y = 1', 'Y = 2'])
axs[0, 0].set_xlabel('X')
axs[0, 0].set_title('p(X, Y)')
for i in range(10):
    axs[0, 0].axvline(i-0.5, color='r', linewidth=1)
axs[0, 0].axhline(1.5, color='r', linewidth=1)

# Plot p(Y)
y_counts = np.bincount(Y)
axs[0, 1].barh([1, 2], y_counts[1:], height=0.5, align='center', color='lightblue')
axs[0, 1].set_yticks([1, 2])
axs[0, 1].set_xlim(0, max(y_counts) * 1.1)
axs[0, 1].set_title('p(Y)')

# Plot p(X)
x_counts = np.bincount(X, minlength=9)
axs[1, 0].bar(range(9), x_counts, color='lightblue')
axs[1, 0].set_xlim(-0.5, 8.5)
axs[1, 0].set_xlabel('X')
axs[1, 0].set_title('p(X)')

# Plot p(X|Y=1)
x_given_y1 = X[Y == 1]
x_given_y1_counts = np.bincount(x_given_y1, minlength=9)
axs[1, 1].bar(range(9), x_given_y1_counts, color='lightblue')
axs[1, 1].set_xlim(-0.5, 8.5)
axs[1, 1].set_xlabel('X')
axs[1, 1].set_title('p(X|Y = 1)')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
