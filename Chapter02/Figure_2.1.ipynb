```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function definition
def f(x1, x2):
    return np.sin(2 * np.pi * x1) * np.sin(2 * np.pi * x2)

# Generate data for the 3D plot
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(0, 1, 100)
x1, x2 = np.meshgrid(x1, x2)
y = f(x1, x2)

# 3D plot
fig = plt.figure(figsize=(18, 6))
ax = fig.add_subplot(131, projection='3d')
ax.plot_surface(x1, x2, y, cmap='viridis')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$y$')
ax.set_title('3D plot of $y(x_1, x_2) = \sin(2\pi x_1) \sin(2\pi x_2)$')

# Generate data for 2D plots
n_samples = 100
x1_samples = np.random.uniform(0, 1, n_samples)
x2_samples = np.random.uniform(0, 1, n_samples)
y_samples = f(x1_samples, x2_samples) + np.random.normal(0, 0.1, n_samples)  # adding Gaussian noise

# Plot when x2 is unobserved
ax = fig.add_subplot(132)
ax.scatter(x1_samples, y_samples, color='red')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$y$')
ax.set_title('Plot with high noise ($x_2$ unobserved)')

# Plot when x2 is fixed to pi/2
x2_fixed = np.pi / 2
y_fixed = f(x1_samples, x2_fixed) + np.random.normal(0, 0.1, n_samples)

ax = fig.add_subplot(133)
ax.scatter(x1_samples, y_fixed, color='red')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$y$')
ax.set_title('Plot with lower noise ($x_2$ fixed to $\pi/2$)')

plt.tight_layout()
plt.show()
```
